```python
# Arithmetic Arranger

def arithmetic_arranger(problems, show_answers=False):
    top_string = ''
    middle_string = ''
    bottom_string = ''
    answer_string = ''
    
    if len(problems) > 5:
        return 'Error: Too many problems.'
    
    for equation in problems:
        num_1, operator, num_2 = equation.split(' ')
        
        try:
            num_1, num_2 = int(num_1), int(num_2)
        except:
            return 'Error: Numbers must only contain digits.'
                                        
        
        spacing = max(len(list(str(num_1))),len(list(str(num_2))))
        
        if spacing > 4:
            return 'Error: Numbers cannot be more than four digits.'
        
        if operator == '+':
            answer = num_1 + num_2
        elif operator == '-':
            answer = num_1 - num_2
        else:
            return "Error: Operator must be '+' or '-'."
    
            
        top_string += f"  {num_1:>{spacing}}    "
        middle_string += f"{operator} {num_2:>{spacing}}    "
        bottom_string += f"{'-'*(2+spacing)}    "
        answer_string += f" {answer:>{spacing+1}}    "
        
    if show_answers == True:
        problems = f"{top_string.rstrip()}\n{middle_string.rstrip()}\n{bottom_string.rstrip()}\n{answer_string.rstrip()}"
    else:
        problems = f"{top_string.rstrip()}\n{middle_string.rstrip()}\n{bottom_string.rstrip()}"

    return problems
```

```python
# Time Calculator

def add_time(start, duration, today = None):
    
    if today is not None:
        days_to_num = {'sunday':0,
                  'monday':1,
                  'tuesday':2,
                  'wednesday':3,
                  'thursday':4,
                  'friday':5,
                  'saturday':6}

        num_to_day = {}

        for day, index in days_to_num.items():
            num_to_day[index] = day.title()
    
        current_day = days_to_num[today.lower()]
        print(current_day)
    
    days = 0
    
    hour, minute = start.split(':')
    minute, meridian = minute.split(' ')
    
    add_hour, add_minute = duration.split(':')
    
    hour, minute, add_hour, add_minute = int(hour), int(minute), int(add_hour), int(add_minute)
    
    if meridian == 'PM' and hour != 12:
        hour += 12
    
    add_minute = add_hour*60 + add_minute
    
    while add_minute > 60:
        hour += 1
        add_minute -= 60
        
    minute += add_minute
    
    while minute >= 60:
        hour += 1
        minute -= 60
    
    final_time = f"{hour}:{minute}"
    
    while hour >= 24:
        hour -= 24
        days += 1
    
    if hour >= 12:
        hour -= 12
        meridian = 'PM'
    else:
        meridian = 'AM'
        
    if hour == 0:
        hour = 12

    if today is not None:
        current_day += days
        print(current_day)
    
        today = num_to_day[current_day % 7]
        today_string = f', {today}'
    else:
        today_string = ''
        
    if days == 1:
        day_string = ' (next day)'
    elif days > 1:
        day_string = f' ({days} days later)'
    else:
        day_string = ''
        
    final_time = f"{hour}:{minute:02} {meridian}{today_string}{day_string}"

    return final_time
```

```python
# Budget App

class Category:
    def __init__(self, name):
        self.name = name
        self.ledger = []
    
    def __str__(self):
        return_line = ''
        
        title_line_list = ['*']*30
        starting_index = int((30 - len(self.name))/2)
        for i in range(starting_index, starting_index + len(self.name)):
            title_line_list[i] = self.name[i-starting_index]
        return_line += ''.join(title_line_list)
        
        for d in self.ledger:
            amount = str(format(d['amount'], '.2f'))
            description = d['description']
            
            line_list = [' ']*30
            
            for i in range(1, len(amount)+1):
                line_list[i*-1] = amount[i*-1]
            
            for i in range(len(description)):
                if i > 22:
                    break
                line_list[i] = description[i]

            line = ''.join(line_list)
            return_line += f'\n{line}'
            
        return_line += f"\nTotal: {self.get_balance()}"

        return return_line
        
    def deposit(self, amount, description = ''):
        self.ledger.append({'amount':amount, 'description':description})
        
    def withdraw(self, amount, description=''):
        if self.check_funds(amount):
            self.ledger.append({'amount':amount*-1, 'description':description})
            return True
        return False
            
    def get_balance(self):
        balance = sum([d['amount'] for d in self.ledger])
        return balance
    
    def transfer(self, amount, Cat):
        if self.withdraw(amount, f'Transfer to {Cat.name}'):
            Cat.deposit(amount, f'Transfer from {self.name}')
            return True
        return False
            
    def check_funds(self, amount):
        if self.get_balance() >= amount:
            return True
        return False
    
def create_spend_chart(categories):
    total = 0
    total_list = []
    for category in categories:
        category_total = 0
        for d in category.ledger:
            if d['amount'] < 0:
                total += d['amount']*-1
                category_total += d['amount']*-1
        total_list.append(category_total)
    percent_list = [int(cat_total/total*100 // 10)*10 for cat_total in total_list]

    return_string = 'Percentage spent by category\n'
    
    for percentile in range(100,-10,-10):
        new_line = f'{percentile:3}|'
        for percent in percent_list:
            if percent >= percentile:
                new_line += ' o '
            else:
                new_line += '   '
        return_string += f"{new_line} \n"
        
    return_string += '    ' + '---'*len(categories) + '-\n'
    
    label_string = ''
    
    for i in range(max([len(category.name) for category in categories])):
        label_string += '    '
        for name in [category.name for category in categories]:
            try:
                label_string += f' {name[i]} '
            except:
                label_string += '   '
        if i != max([len(category.name) for category in categories])-1:
            label_string += ' \n'
        else:
            label_string += ' '
                
    return_string += label_string  
    
    return return_string
```

```python
# Polygon Area Calculator

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"
        
    def set_width(self, new_width):
        self.width = new_width

    def set_height(self, new_height):
        self.height = new_height
    
    def get_area(self):
        return self.width * self.height
    
    def get_perimeter(self):
        return 2*self.width + 2*self.height
    
    def get_diagonal(self):
        return ((self.width ** 2 + self.height ** 2) ** .5)
    
    def get_picture(self):
        picture = ''
        
        if (self.width > 50) or (self.height > 50):
            return "Too big for picture."
        
        for _ in range(self.height):
            picture += f"{'*'*self.width}\n"

        return picture
    
    def get_amount_inside(self, rec_0):
        if (rec_0.height > self.height) or (rec_0.width > self.width):
            return 0
        else:
            extra_width = self.width
            extra_height = self.height
            width_fits = 0
            height_fits = 0
            
            while extra_width >= rec_0.width:
                width_fits += 1
                extra_width -= rec_0.width
            while extra_height >= rec_0.height:
                height_fits += 1
                extra_height -= rec_0.height
                
        return height_fits * width_fits
                
                
class Square(Rectangle):
    def __init__(self, side):
        self.height = side
        self.width = side
        
    def __str__(self):
        return f"Square(side={self.width})"
    
    def set_side(self, new_side):
        self.height = new_side
        self.width = new_side
        
    def set_width(self, new_width):
        self.width = new_width
        self.height = new_width

    def set_height(self, new_height):
        self.height = new_height
        self.width = new_height
```

```python
# Probability Calculator

import random

class Hat:
        
    def __init__(self, **kwargs):
        d = {key:value for key, value in kwargs.items()}

        self.contents = [key for key, value in d.items() for _ in range(value)]
    
    def draw(self, num_balls):
        if num_balls > len(self.contents):
            num_balls = len(self.contents)
            
        drawn_balls = random.sample(self.contents, num_balls)
        
        return drawn_balls
            
        
def experiment(hat, expected_balls, num_balls_drawn, num_experiments):
    M = 0
    
    for _ in range(num_experiments):
        result = hat.draw(num_balls_drawn)
        print(result)
        
        successes = [result.count(color) >= amount for color,amount in expected_balls.items()]
        print(successes)
        
        if False not in successes:
            M += 1
            
    return M/num_experiments
```
